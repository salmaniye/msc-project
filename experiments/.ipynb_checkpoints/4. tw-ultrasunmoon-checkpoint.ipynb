```python
import csv
import tweepy
import configparser

# read config
config = configparser.ConfigParser()
config.read('config.ini')

# Oauth keys
consumer_key = config['twitter']['api_key']
consumer_secret = config['twitter']['api_key_secret']

access_token = config['twitter']['access_token']
access_token_secret = config['twitter']['access_token_secret']

bearer_token = 'AAAAAAAAAAAAAAAAAAAAAD5SegEAAAAAGbYKP2c5xC%2BcUMU4hvkxiRThhhw%3DKjPc1ClSbGmwoB6JlSxupf54rNddyHLndzxA7tVdfDvEsWVpOT'

# Authentication with Twitter
auth = tweepy.OAuth1UserHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)
client = tweepy.Client(bearer_token = bearer_token, wait_on_rate_limit=True)
```

```python
# Pokemon Sun and Moon (2016)
# tweets = client.search_all_tweets(query = '("Pokemon Sun" OR "Pokemon Moon" OR "Pokemon Sun & Moon" OR #PokemonSunMoon OR #SunMoon) -is:retweet lang:en', max_results = 500,
#          start_time='2016-02-26T00:00:00Z', end_time='2016-11-18T00:00:00Z')
tweets = []
for response in tweepy.Paginator(client.search_all_tweets, query='("Pokemon Sun" OR "Pokemon Moon" OR "Pokemon Sun & Moon" OR #PokemonSunMoon OR #SunMoon) -is:retweet lang:en',
        start_time='2016-02-26T00:00:00Z', end_time='2016-11-18T00:00:00Z', 
        tweet_fields="created_at", max_results=500, limit=500):
    tweets.append(response)

with open('pokemon_sunmoon.csv', 'w') as f:
    csv_writer = csv.DictWriter(f, fieldnames=('tweet id', 'text','created at'))
    csv_writer.writeheader()
    for i in range(len(tweets)):
        for tweet in tweets[i].data:
            row = {'tweet id' : tweet.id,'text': tweet.text,'created at' : tweet.created_at}
            csv_writer.writerow(row)
```
