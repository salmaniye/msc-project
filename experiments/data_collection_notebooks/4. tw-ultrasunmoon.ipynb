```python
# Pokemon Ultra Sun and Ultra Moon (2017)
# 06/06/2017 - 17/11/2017
# tweets = client.search_all_tweets(query = '() -is:retweet lang:en', max_results = 500,
#          start_time='2017-06-06T00:00:00Z', end_time='2017-11-18T00:00:00Z')
tweets = []
for response in tweepy.Paginator(client.search_all_tweets, query='("Pokemon Ultra Sun and Ultra Moon" OR "Pokemon Ultra Sun" OR "Pokemon Ultra Moon" OR #PokemonUltraSunMoon OR #PokemonUltraSunandUltraMoon) -is:retweet lang:en',
        start_time='2017-06-06T00:00:00Z', end_time='2017-11-18T00:00:00Z', 
        tweet_fields="created_at", max_results=500, limit=500):
    tweets.append(response)

with open('pokemon_ultrasunmoon.csv', 'w') as f:
    csv_writer = csv.DictWriter(f, fieldnames=('tweet id', 'text','created at'))
    csv_writer.writeheader()
    for i in range(len(tweets)):
        for tweet in tweets[i].data:
            row = {'tweet id' : tweet.id,'text': tweet.text,'created at' : tweet.created_at}
            csv_writer.writerow(row)
```
