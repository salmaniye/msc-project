```python
# Pokemon X & Y (2013)
# tweets = client.search_all_tweets(query = '#PokemonXY', max_results = 500,
#          start_time='2013-01-23T00:00:00Z', end_time='2013-10-12T00:00:00Z')
tweets = []
for response in tweepy.Paginator(client.search_all_tweets, query='("Pokemon X" OR "Pokemon Y" OR "Pokemon X&Y" OR #PokemonXY OR #PokemonX OR #PokemonY) -is:retweet lang:en',
        start_time='2013-01-23T00:00:00Z',  end_time='2013-10-13T00:00:00Z', 
        tweet_fields="created_at", max_results=500, limit=500):
    tweets.append(response)

with open('pokemon_xy.csv', 'w') as f:
    csv_writer = csv.DictWriter(f, fieldnames=('tweet id', 'text','created at'))
    csv_writer.writeheader()
    for i in range(len(tweets)):
        for tweet in tweets[i].data:
            row = {'tweet id' : tweet.id,'text': tweet.text,'created at' : tweet.created_at}
            csv_writer.writerow(row)
```
