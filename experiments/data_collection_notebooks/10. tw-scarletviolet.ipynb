```python
# Pokemon Scarlet Violet (2022)
# tweets = client.search_all_tweets(query = '() -is:retweet lang:en', max_results = 500,
#          start_time='2022-02-27T00:00:00Z', end_time='2022-01-29T00:00:00Z')

tweets = []
for response in tweepy.Paginator(client.search_all_tweets, query='(#PokemonScarletViolet OR #PokemonScarlet OR #PokemonViolet OR "Pokemon Scarlet" OR "Pokemon Violet" OR "Pokemon Scarlet and Violet" OR "Pokemon Scarlet & Violet") -is:retweet lang:en',
        start_time='2022-02-27T00:00:00Z', end_time='2022-08-08T00:00:00Z', 
        tweet_fields="created_at", max_results=500, limit=5000):
    tweets.append(response)

with open('pokemon_sv.csv', 'w') as f:
    csv_writer = csv.DictWriter(f, fieldnames=('tweet id', 'text','created at'))
    csv_writer.writeheader()
    for i in range(len(tweets)):
        for tweet in tweets[i].data:
            row = {'tweet id' : tweet.id,'text': tweet.text,'created at' : tweet.created_at}
            csv_writer.writerow(row)
```
