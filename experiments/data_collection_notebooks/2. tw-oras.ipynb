```python
# Pokemon ORAS (2014)
# tweets = client.search_all_tweets(query = "Pokemon Omega Ruby" OR "Pokemon Alpha Sapphire" OR "Pokemon ORAS" OR #PokemonORAS OR #ORAS) -is:retweet lang:en', max_results = 500,
#          start_time='2014-05-07T00:00:00Z', end_time='2014-11-21T00:00:00Z')
tweets = []
for response in tweepy.Paginator(client.search_all_tweets, query='("Pokemon Omega Ruby" OR "Pokemon Alpha Sapphire" OR "Pokemon ORAS" OR #PokemonORAS OR #ORAS) -is:retweet lang:en',
        start_time='2014-05-07T00:00:00Z',  end_time='2014-11-22T00:00:00Z', 
        tweet_fields="created_at", max_results=500, limit=500):
    tweets.append(response)

with open('pokemon_oras.csv', 'w') as f:
    csv_writer = csv.DictWriter(f, fieldnames=('tweet id', 'text','created at'))
    csv_writer.writeheader()
    for i in range(len(tweets)):
        for tweet in tweets[i].data:
            row = {'tweet id' : tweet.id,'text': tweet.text,'created at' : tweet.created_at}
            csv_writer.writerow(row)
```
